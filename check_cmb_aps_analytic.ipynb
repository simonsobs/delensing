```python
import numpy as np, prjlib, basic, plottools as pl
from matplotlib.pyplot import *
```

```python
p = prjlib.analysis_init()
```

```python
l, Nl = prjlib.nlspecs(t='la',freqs=['93','145','225','280'])
l, N2 = prjlib.nlspecs(t='la',freqs=['93','145'])
```

```python
l, nl = prjlib.nlspecs(t='sa',freqs=['93','145','225','280'])
l, n2 = prjlib.nlspecs(t='sa',freqs=['93','145'])
l, n3 = prjlib.nlspecs(t='sa',freqs=['93','145','225'])
```

```python
pl.plot_1dstyle(ylog=True,xmin=1,xmax=2048,ymin=4e-6,ymax=1e-3,ylab=r'$C^{EE}_L$ [$\mu K^2$]')
plot(l,Nl['all'],label='all')
plot(l,N2['all'],label='93+145')
plot(l,p.lEE*prjlib.Tcmb**2,color='k')
legend()
```

```python
pl.plot_1dstyle(ylog=True,xmin=1,xmax=2048,ymin=1e-7,ymax=1e-3,ylab=r'$C_L^{BB}$ [$\mu K^2$]')
for freq in ['93','145','225','280']:
    plot(l,nl[freq],lw=.5)
plot(l,N2['all'],label='93+145 (LAT)',color='r',ls='--')
plot(l,n2['all'],label='93+145 (SAT)',color='r')
plot(l,n3['all'],label='93+145+225 (SAT)',color='m')
plot(l,nl['all'],label='93+145+225+280 (SAT)',color='b')
plot(l,p.lBB*prjlib.Tcmb**2,'k--',lw=.5)
legend()
```
