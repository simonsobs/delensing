```python
import healpy as hp, numpy as np, prjlib, warnings, tqdm
from matplotlib.pyplot import *
warnings.filterwarnings("ignore")
H = {}
D = '/project/projectdirs/sobs/sims/mss-0001/mss-001_madam_v1.0/healpix/'
```

```python
H['la'] = 0.
for s in ['0','1','2','3']:
    for lt in tqdm.tqdm(['2','3','4','5']):
        H['la'] += hp.fitsfunc.read_map(D+'lat_s21_wide_LT'+lt+'_f093_4way_set0'+s+'_hits.fits')
```

```python
H['sa'] = 0.
for s in ['0','1','2','3']:
    for lt in tqdm.tqdm(['1','2']):
        H['sa'] += hp.fitsfunc.read_map(D+'sat'+lt+'_s21_deep_ST'+lt+'_f093_4way_set00_hits.fits')
```

```python
for t in ['la','sa']:
    hp.mollview(H[t]/np.max(H[t]),min=0,max=1,title='Hitcount, '+t)
```
