```python
import numpy as np, prjlib, plottools as pl
from matplotlib.pyplot import *
```

```python
pla, fla, __ = prjlib.analysis_init(t='la',freq='coadd')
pco, fco, __ = prjlib.analysis_init(t='co',freq='coadd')
pl.plot_1dstyle(xmin=2,xmax=150,ymin=.3,ymax=.5,ylab=r'$1-(\rho^{BB})^2$')
l, bb, BB, Bb = np.loadtxt(fla.delens['ideal'].scl,unpack=True,usecols=(0,1,2,3))
plot(l,1.-Bb**2/bb/BB,ls='--',label='LAT E-modes (ideal)')
l, bb, BB, Bb = np.loadtxt(fco.delens['ideal'].scl,unpack=True,usecols=(0,1,2,3))
plot(l,1.-Bb**2/bb/BB,ls='--',label='LAT+SAT E-modes (ideal)')
l, bb, BB, Bb = np.loadtxt(fco.delens['wiener'].scl,unpack=True,usecols=(0,1,2,3))
plot(l,1.-Bb**2/bb/BB,label='LAT+SAT E-modes (wiener)')
legend()
savefig('fig_rhobb.png')
```
