```python
import numpy as np, prjlib, basic, plottools as pl
from matplotlib.pyplot import *
```

```python
lcl  = basic.aps.read_cambcls('/project/projectdirs/sobs/delensing/cls/ffp10_lensedCls.dat',2,4096,4,bb=True)
```

```python
pl.plot_1dstyle(ylog=True,xmin=1,xmax=2048,ymin=4e-6,ymax=1e-3)
l, Nl = prjlib.nlspecs(t='la',freqs=['93','145','225','280'])
N2 = (Nl['93']*Nl['145'])/(Nl['93']+Nl['145']+1e-30)
plot(l,Nl['mv'],label='all')
plot(l,N2,label='93+145')
plot(l,lcl[1,:],color='k')
legend()
```

```python
ep = 1e-30
pl.plot_1dstyle(ylog=True,xmin=1,xmax=2048,ymin=1e-7,ymax=1e-3)
l, Nl = prjlib.nlspecs(t='sa',freqs=['93','145','225','280'])
l, n2 = prjlib.nlspecs(t='sa',freqs=['93','145'])
l, n3 = prjlib.nlspecs(t='sa',freqs=['93','145','225'])
for freq in ['93','145','225','280']:
    plot(l,Nl[freq],lw=.5)
plot(l,Nl['mv'],label='all (SAT)',color='b')
plot(l,N2,label='93+145 (LAT)',color='r',ls='--')
plot(l,n2['mv'],label='93+145 (SAT)',color='r')
plot(l,n3['mv'],label='93+145+225 (SAT)',color='m')
plot(l,lcl[2,:],'k--',lw=.5)
legend()
```
