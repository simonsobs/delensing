```python
import numpy as np, prjlib
from matplotlib.pyplot import *
```

```python
pla, fla, __ = prjlib.analysis_init(t='la',freq='coadd')
pco, fco, __ = prjlib.analysis_init(t='co',freq='coadd')
xlim(2.,200.)
ylim(.2,.5)
grid(True)
l, bb, BB, Bb = np.loadtxt(fla.delens['ideal'].scl,unpack=True,usecols=(0,1,2,3))
plot(l,1.-Bb**2/bb/BB,label='la')
l, bb, BB, Bb = np.loadtxt(fco.delens['ideal'].scl,unpack=True,usecols=(0,1,2,3))
plot(l,1.-Bb**2/bb/BB,label='co')
legend()
```

```python
D = '/project/projectdirs/sobs/delensing/delensb/derived/'
xlabel('$\ell$')
ylabel('Residual $BB$')
xlim(2,200)
ylim(.2,.5)
for ms,c in [('simple','m'),('wiener','g')]:
    b, bb, LL, bL = np.loadtxt(D+'scl_'+ms+'_Eco_dl20-2048.dat',unpack=True,usecols=(0,1,2,3))
    res = 1.-bL**2/bb/LL
    plot(b,res,c+'-',label='masked')
b, bb, LL, bL = np.loadtxt(D+'scl_ideal_Ela_dl20-2048.dat',unpack=True,usecols=(0,1,2,3))
res = 1.-bL**2/bb/LL
plot(b,res,'--',color='orange')
b, bb, LL, bL = np.loadtxt(D+'scl_ideal_Eco_dl20-2048.dat',unpack=True,usecols=(0,1,2,3))
res = 1.-bL**2/bb/LL
plot(b,res,'-',color='orange')
#savefig('fig_delens.png')
```
